```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('ligands_allfolds_e273_GCN_epochs_1365.csv')
data
```

```python
data.loc[(data['fold'] ==1) & (data['set'] == 'test')]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
r2 = []
mae = []
rmse = []
```

```python
for fold in range(10):
    df = data.loc[(data['fold'] ==fold) & (data['set'] == 'test')]
    y = df['ee']
    pred = df['Prediction']
    r2.append(r2_score(y,pred))
    mae.append(mean_absolute_error(y,pred))
    rmse.append(sqrt(mean_squared_error(y,pred)))
```

```python

Metrics = ['R2', 'MAE', 'RMSE']
x_pos = np.arange(len(Metrics))
CTEs = [np.mean(r2), np.mean(mae), np.mean(rmse)]
error = [np.std(r2), np.std(mae), np.std(rmse)]

fig, ax = plt.subplots(figsize =(6, 4))
bars = ax.bar(x_pos, CTEs, yerr=error, align='center', color = 'navy', alpha=.5, ecolor='black', capsize=10)
ax.set_ylabel('Value', fontweight ='bold', fontsize = 12)
ax.bar_label(bars)
ax.set_xticks(x_pos)
ax.set_xticklabels(Metrics, fontweight ='bold' ,fontsize = 10)
ax.yaxis.grid(False)

plt.tight_layout()
plt.show()
```
