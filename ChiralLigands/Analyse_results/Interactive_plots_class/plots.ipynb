```python
import pandas as pd
import plotly.express as px
import molplotly
```

```python
data = pd.read_csv('/Users/ed/Documents/PhD_Documents/Annual Review/1st year/Code/ChiralLigands/Analyse_results/Interactive_plots_class/ligands_allfolds_e47_GCN_epochs_235.csv')
data['logits'] = -data['logits']
data = data.loc[data['fold']==9]
data.sample(10)
```

```python
fig_scatter = px.scatter(data,
                         x="logits",
                         y="ee",
                         color='set',
                         labels={'logits': '-Logits',
                                 'ee': 'e.e',
                                 'Boron_reagent': 'Boron reagent'},
                         template = "none",
                         width=900,
                         height=600)


y = data["ee"].values
fig_scatter.add_shape(
    type="line", line=dict(dash='dash'),
    x0=0, y0=y.min(),
    x1=0, y1=y.max()
)

x = data["logits"].values
fig_scatter.add_shape(
    type="line", line=dict(dash='dash'),
    x0=x.min(), y0=50,
    x1=x.max(), y1=50
)


app_scatter = molplotly.add_molecules(fig = fig_scatter,
                                      df = data,
                                      smiles_col = ['Ligand'], 
                                      title_col = 'index',
                                      show_coords = True,
                                      caption_transform={'Logits': lambda x: f"{x:.2f}",
                                                                       'e.e': lambda x: f"{x:.2f}"
                                                                       },
                                      color_col='set',
                                      alpha = 0.4,
                                      mol_alpha= 0.7
                                      )

# change the arguments here to run the dash app on an external server and/or change the size of the app!
app_scatter.run_server(mode='inline', port=2135, height=1000)
```
